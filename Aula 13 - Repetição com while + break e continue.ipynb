## 0 - Operadores de atribuição

Usando repetições, é comum ter que mudar os valores nas variáveis.

O código abaixo aumenta em uma unidade o valor associado a n.

```python
n = 10
n = n + 1
n
```

Podemos fazer o mesmo usando +=

```python
n = 10
n += 1
n
```

Podemos usar a forma <b>operador=</b> para fazer atribuições de um jeito mais curto.

```python
n = 3
n -= 1
n
```

```python
n = 4
n /= 2
n
```

```python
n = 4
n %= 2
n
```

```python
n = 4
n //= 2
n
```

Programador tende a escrever o tanto que precisa de código, nada mais.

## 1 - Repetição com while

O while é usado para repetir instruções enquanto uma condição for satisfeita.

```python
n = 0
while n < 3:
    print(n)
    n += 1
```

Como o while é uma estrutura (de repetição), tudo dentro dele tem indentação.

E cuidado com a condição do while, senão ele vai executar "pra sempre".

Esse é o chamado loop infinito.

```python
while 10 > 9:
    print('a', end = '')
```

## 2 - Condições mais complexas

Tudo o que usamos em estruturas de decisão com if pode ser usado na condição do while.

```python
n = 0
while 1 != 2 and n <= 3:
    print(n, end = ' ')
    n += 1
```

Inequações:

```python
n = 0
while n < 3 < 5:
    print(n, end = ' ')
    n += 1
```

## 3 - Comando break 

Se você quiser sair do while enquanto ele está sendo executado, basta usar o break.

Esse comando vai ignorar o que está abaixo dele (dentro do while) e vai sair da repetição.

```python
n = 10
while n > 0:
    if n == 15: 
        break
    print(n, end = ' ')
    n += 1
print('saiu do while e n vale', n)
```

## 4 - Comando continue

Você pode precisar de algo como o break, mas que volte ao loop após pular o código abaixo.

Fazemos isso com o continue.

```python
n = 0
while n < 5:
    n += 1
    if n == 3:
        continue
    print(n, end = ' ')
```

Quando o if é satisfeito, o continue é executado.

Por causa disso ele pula o print quando n é 3. A impressão de 4 e 5 mostra que ele volta ao loop.

## 5 - Else

Podemos usar o else pra capturar o que sair do while quando sua condição não for satisfeita.

```python
n = 1
while n < 0:
    print('n < 0')
    n += 1
else:
    print('Não entrou no while')
```

```python
n = -1
while n < 0:
    print('n < 0')
    n += 1
else:
    print('Saiu do while, pois n =', n)
```

Lembre-se que o else só é ativado quando a condição do while não é satisfeita.

Então o else não é executado se você sair do while com um break (tente você mesmo).
